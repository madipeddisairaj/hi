## **CODE**

```python
import math
x = 3
y = 2
n = 0.01 
eps = 0.000001 
max_iters = 100
iters= 0 
delta_x=delta_y=1
def grad(x,y):
 dx = 6 * x
 dy = -5 * math.exp(-y)
 return dx,dy
while max(abs(delta_x),abs(delta_y)) > eps and iters < max_iters:
 prev_x = x 
 prev_y = y 
 delta_x,delta_y=grad(prev_x,prev_y)
 delta_x=-n * delta_x
 delta_y=-n * delta_y
 x = x+delta_x 
 y = y+delta_y 
 iters= iters+1
 if iters<=2 :
   print('At iteration',iters,': x =',x,'y =',y,)   
print("\nThe local minimum occurs at", x,y)
```
